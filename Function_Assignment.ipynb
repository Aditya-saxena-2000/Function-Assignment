Q.1 What is the difference between a function and a method in Python?

Ans. In Python, both functions and methods are pieces of code with a specific task. The difference usually lies in the following:

**Function:**
 A function is a block of code that may be invoked from anywhere in your program. It is defined with the keyword def.

 A method is a function that is associated with an object -an instance of a class.
It is defined inside a class and it can access an object's attributes and other methods.

```python
#Example:
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello,", self.name)
```

```python
person = Person("Aditya")
person.greet()
```

Functions are defined outside of classes whereas methods will be inside the class.

Methods are associated with objects and can have access to their data; functions are not.

**Calling:**
Functions are directly called by their name whereas methods are called on an object by dot notation-for example, object.method().

Q.6  Explain the concept of generators in Python and how they are defined.

Ans. Generators provide a simple way to create iterators using functions and are defined with the yield keyword in Python instead of the return value.

**Important facts:**

**Yield Keyword**

**Yield vs. Return:**

 Unlike the return statement, which terminates the function, yield returns a value and saves the state of the function, then the execution will continue from the point at which the last yield executed.

**Multiple Yields:**

 Functions can yield multiple values over time; it generates a series of results.

**Lazy Evaluation:**

On-demand data: Values are produced one at a time, and only if and when they're required to be produced, which means you never have to produce all values upfront and simultaneously store them somewhere in memory.

**Memory efficiency:**

 Due to these reasons, generator objects themselves use minimal memory. Properties Iterables Data is returned in a format usable in an iterable context, so that the sequence can be iterated over, much like one might iterate over a list or tuple.

  **Advantages**

**Efficient Usage of Memory:**

 Generators pull things from memory, rather than storing an entire sequence in memory; thus, they are very memory-efficient for large datasets or infinite sequences.

**Simplified Syntax:**

 Making use of yield simplifies the creation of iterators. It's much shorter and more readable.

```python
def countdown(n):
  while n > 0:
    yield n
    n-=1
```

```python
# Creating a Generator
gen = countdown(5)
```

```python
#Using the Generator
for num in gen:
  print(num)
```

Q.7 What are the advantages of using generators over regular functions?

Ans. There are several advantages of using generators over regular functions mentioned bellow:

1.**Memory Efficiency:**

One of the major advantages of generators is a factor of memory efficiency. Since a generator yields its values one by one, there is no need to store all values in memory at once. In particular, this comes in handy when working with voluminous data, because doing so reduces the chance of running out of memory.

2.**Readability:**

Generating can make your code more readable. Since the process of iteration is split into smaller pieces, you sometimes could write more concise code that's easier to follow.

3.**Speed:**

 Generators also can be faster compared to regular loops, especially operating with large sets of data. This is because it doesn't have memory and processing overhead for building a list or other data structure to hold results.

The following is an example of generators that highights the memory efficiency and readability gains:

```python
def first_millon_numbers():
  for i in range(1000000):
    yield i
```

```python
gen = first_millon_numbers()
print(sum(gen))
```

The first_million_numbers generator produces the first million numbers, yielding them one after another. The sum function consumes the generator computing a summation over all values produced. Since the generator creates it values one at a time, its values don't all have to be in memory at one time.

**Conclusion**


Generators are one of the wonderful and versatile features in the Python programming language. Basically, yielding values one at a time has its advantages compared with traditional loops in terms of memory efficiency and readability, among others. A lambda function can take any number of arguments, but can only have one expression.

**Syntax**
lambda arguments:expression


The expression is executed and the result is returned:

Lambda functions can take any number of arguments:

```python
#Example: Multiply argument a with argument b and return the result:
x = lambda a, b : a* b
print(x(100,80))
```

Q.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans. Well, map(), reduce(), and filter() are some of the built-in functions in Python that allow for the processing of iterables-like lists, tuples, etc.-in a functional way. Now, let's identify the differences between them.

**map():**

The purpose of this function is to apply a given function to each element of an iterable and return a new iterable containing the transformed values.

**Syntax:** map(function, iterable)


```python
#Example
numbers = [1,2,3,4,5]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

**filter():**

**Filter:** It takes a given function and applies to all items of an iterable. The filtered items are then used to generate a new iterable.

**Syntax:** filter(function, iterable)

```python
#Example
numbers = [1,2,3,4,5]
evens = filter(lambda x: x%2==0, numbers)
print(list(evens))
```

**reduce():**


Accumulates elements of an iterable on the left into one value by applying a given function cumulatively.

**Syntax:** reduce(function, iterable, [initializer])


```python
#Example
from functools import reduce
numbers = [1,2,3,4,5]
sum_of_numbers = reduce(lambda x,y: x+y, numbers)
print(sum_of_numbers)
```

**Main Differences:**


**Output:** map() returns a new iterable containing values which are mapped to; filter() returns a new iterable containing only filtered values, while reduce() returns one value.


**Function:**
map() applies a function to each element independently, filter() includes an element or not based on a function, and reduce() uses a function to combine elements cumulatively.

**Use Cases:** map() is useful when a transformation needs to take place on the data, filter() is useful when one needs to extract elements based on specific criteria, and reduce() is useful when aggregating data in some way.

```python
#Ans.

class FibonacciIterator:
  def __init__(self,max_terms): # Fix the spelling of the constructor.
    self.max_terms = max_terms
    self.current_term = 0
    self.previous_term = 0
    self.next_term = 1
  def __iter__(self): # Fix the spelling of the iterator.
    return self
  def __next__(self):
        if self.current_term >= self.max_terms:
            raise StopIteration
        current_term = self.next_term
        self.next_term = self.previous_term + current_term
        self.previous_term = current_term
        self.current_term += 1
        return current_term
```

Q.7  Implement a generator function that reads a file line by line and yields each line as a string.

```python
#Ans.
#Ans.
def read_file_line_by_line(file_path):
  try:
    with open(file_path, 'r') as file:
      for line in file: # Added indentation here
        yield line.strip() # Remove leading/trailing whitespace characters
  except FileNotFoundError:
    print(f"File not found: {file_path}. Please check the file path and try again.")
```

```python
#Example usage
# Assuming 'example.txt' is a file with some text.
for line in read_file_line_by_line('example.txt'):
  print(line) # Added indentation here
```
